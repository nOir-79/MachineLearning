```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', as_frame = False)
```

```python
X,y = mnist.data, mnist.target
```

```python
X_train, X_test, y_train, y_test = X[:60000],X[60000:],y[:60000],y[60000:]
```

```python
def shift_image(image, dx=1, dy=1):
    reshaped_image = image.reshape(28,28)
    shifted_image = np.roll(reshaped_image, shift=dx, axis=1)  # Horizontal shift
    shifted_image = np.roll(shifted_image, shift=dy, axis=0)  # Vertical shift
    
    # Zero out the parts that are wrapped around
    if dx > 0:
        shifted_image[:, :dx] = 0
    elif dx < 0:
        shifted_image[:, dx:] = 0
    
    if dy > 0:
        shifted_image[:dy, :] = 0
    elif dy < 0:
        shifted_image[dy:, :] = 0
    
    return shifted_image.flatten()
```

```python
X_train_shifted_right = np.array([shift_image(img, dx=2, dy=0) for img in X_train])
```

```python
X_train_shifted_left = np.array([shift_image(img, dx=-2, dy=0) for img in X_train])
```

```python
X_train_shifted_top = np.array([shift_image(img, dx=0, dy=2) for img in X_train])
```

```python
X_train_shifted_bottom = np.array([shift_image(img, dx=2, dy=-2) for img in X_train])
```

```python
indices = np.arange(len(X_train))
np.random.shuffle(indices)
```

```python
indices = np.arange(len(X_train))
np.random.shuffle(indices)
scrambled_X_train_shifted_right = X_train_shifted_right[indices]
scrambled_y_train_shifted_right = y_train[indices]
```

```python
indices = np.arange(len(X_train))
np.random.shuffle(indices)
scrambled_X_train_shifted_left = X_train_shifted_left[indices]
scrambled_y_train_shifted_left = y_train[indices]
```

```python
indices = np.arange(len(X_train))
np.random.shuffle(indices)
scrambled_X_train_shifted_top = X_train_shifted_top[indices]
scrambled_y_train_shifted_top = y_train[indices]
```

```python
indices = np.arange(len(X_train))
np.random.shuffle(indices)
scrambled_X_train_shifted_bottom = X_train_shifted_bottom[indices]
scrambled_y_train_shifted_bottom = y_train[indices]
```

```python
shifted_X_train = np.concatenate([X_train,scrambled_X_train_shifted_left,scrambled_X_train_shifted_right,scrambled_X_train_shifted_top,scrambled_X_train_shifted_bottom])
```

```python
shifted_y_train = np.concatenate([y_train,scrambled_y_train_shifted_left,scrambled_y_train_shifted_right,scrambled_y_train_shifted_top,scrambled_y_train_shifted_bottom])
```

```python
indices = np.arange(len(shifted_X_train))
np.random.shuffle(indices)
cleaned_X_train= shifted_X_train[indices]
cleaned_y_train= shifted_y_train[indices]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=7,weights = 'distance')
```

```python
knn.fit(cleaned_X_train,cleaned_y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(knn, cleaned_X_train, cleaned_y_train, cv = 3, scoring = "accuracy")
```
